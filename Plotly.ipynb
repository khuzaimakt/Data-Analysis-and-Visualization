```python
import numpy as np
import pandas as pd
import plotly as pt
import plotly.offline as pto
import plotly.graph_objects as go
import plotly.express as px
```

```python
df = pd.read_csv("Top_Daily.csv")
df.drop(columns=["Unnamed: 0"], errors="ignore", inplace=True)
df[df["Region"] == "Pakistan"].tail()
```

```python
pkdf = df[df["Region"] == "Pakistan"].copy()
pkdf.drop(columns=[
    "Track Primary Artist", 
    "Track Link", 
    "Artist Genres", 
    "Region"], errors="ignore", inplace=True)

pkdf
```

```python
pkdf.corr()
```

```python
fig = go.Figure(
    data=go.Bar(x=pkdf["Track Name"], y=(pkdf["Duration(MS)"] / (1000 * 60))),
    layout_title_text="Duration in minutes by song titles"
)

fig.show()
```

```python
pkdf["Duration(MIN)"] = (pkdf["Duration(MS)"] / (1000 * 60))
```

```python
fig = px.scatter(pkdf,
    x="Track Popularity", y="Duration(MIN)",
)

fig.show()
```

```python
fig = px.line(pkdf,
             x="Loudness", y="Energy")
fig.show()
```

```python
fig = px.scatter(df, y="Duration(MS)", x="Track Popularity",
                 color="Region", symbol="Region")
fig.update_traces(marker_size=10)
fig.show()
```

```python
aggdf = df.groupby("Region", as_index=False).mean()
aggdf["Duration(MIN)"] = aggdf["Duration(MS)"] / (1000 * 60)
fig = px.pie(aggdf, names="Region", values="Duration(MIN)")
fig.show()
```
