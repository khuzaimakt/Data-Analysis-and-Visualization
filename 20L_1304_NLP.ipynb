```python
import numpy as np 
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
import string
import nltk
import re
import math
from string import punctuation
```

```python
def Words_Extract(df,words):
    for tweet in df['tweet']:
        for word in tweet:
            if word not in words:
                words.append(word)
```

```python
def generate_N_grams(words,ngram):
  temp=zip(*[words[i:] for i in range(0,ngram)])
  ans=[' '.join(ngram) for ngram in temp]
  return ans
```

```python
def BoW(df,words):
    vectors = list()
    for tweet in df['tweet']:
        wordfreq=dict()
        tweet = " ".join(tweet)
        print(tweet)
        for word in words:
            
            wordfreq[word]=tweet.count(word)
        vectors.append(wordfreq.values())

    bow=dict()
    for idx, i in enumerate(vectors):
        bow[f'bow_{idx}'] = i

    df['Bow']= bow.values()
```

```python
def TF_IDF(df):
    TFIDF=list()
    N=10

    for tweet in df['tweet']:
        
        TF_IDF=dict()
        j=0
        idf=0
        for w in tweet:
            j+=1
        
        for word in tweet:
            for t in df['tweet']:
                if word in t:
                    idf+=1
            TF_IDF[word]=max(0,((tweet.count(word))/j) * (1 + math.log(N/idf))) #relu fucntion
                    

        
        TFIDF.append(TF_IDF.values())

    terminv=dict()
    for idx, i in enumerate(TFIDF):
        terminv[f'tfidf_{idx}'] = i

    df['TF_Idf']= terminv.values()
```
